```java

public class SoccerDynamic {
    private String name;// private stings in order to make it acssessed only through the classs SoccerDynamic
    private String league;
    private String position;
    
    public SoccerDynamic () { // sets a constructor with out parameters INSTANCE VAIRBLES
        this.name = "Messi";  // using setter with static string
        this.league = "MLS"; // sets intial values 
        this.position = "LW";
    }
   
    public SoccerDynamic(String name, String league, String position) { // 3 Parameter constructor. CONSTRUCTORS 
        this.name = name;
        this.league = league;
        this.position = position;  // this.xx =xxx; is to set and initialize the variable as a parameter
    }
    
    public void setName(String name) { // this is a method which allows you to change the name instance variable. 
        this.name = name;    
    }
    public void setLeague(String league) {
        this.league = league;   
    }
    public void setPosition(String position) { 
        this.position = position;    
    }

    public void displayInfo() { // this is used to print the players name such as messi
        System.out.println("name:" + name);
        System.out.println("league:" + league);
        System.out.println("position:" + position);

    }
    public static void main(String[] args) {  // creates a statics class and which allows for parameters. 
        SoccerDynamic soccer1 = new SoccerDynamic(); // use of a parameterless constructor
        SoccerDynamic soccer2 = new SoccerDynamic("Ronaldo", "Saudi Pro", "ST"); // creates new data within a new constructors 

        System.out.println("Info:"); // accessing getter
        System.out.println("Player 1:"); // printing player 1 
        soccer1.displayInfo();
        System.out.println(); 

        System.out.println("Player 2:"); // printing player 2 
        soccer2.displayInfo();
        System.out.println();
    }
}


public class SoccerDynamic { // name the first letter of class as capitalized, note camel case
    // instance variable have access modifier (private is most common), data type, and name
    private String name;
    private String league;
    private String position;
    // constructor signature 1, public and zero arguments, constructors do not have return type
    public SoccerDynamic () {  // 0 argument constructor
        this.name = "Messi";  // using setter with static string
        this.league = "MLS";
        this.position = "LW";
    }
    // constructor signature, public and one argument
    public SoccerDynamic(String name, String league, String position) { // 1 argument constructor
        this.name = name;
        this.league = league;
        this.position = position;   // using setter with local variable passed into constructor
    }
    // setter/mutator, setter have void return type and a parameter
    public void setName(String name) { // setter
        this.name = name;     // instance variable on the left, local variable on the right
    }
    public void setLeague(String league) { // setter
        this.league = league;     // instance variable on the left, local variable on the right
    }
    public void setPosition(String position) { // setter
        this.position = position;     // instance variable on the left, local variable on the right
    }
    // getter/accessor, getter used to return private instance variable (encapsulated), return type is String
    public void displayInfo() {
        System.out.println("name:" + name);
        System.out.println("league:" + league);
        System.out.println("position:" + position);

    }
    public static void main(String[] args) {  
        SoccerDynamic soccer1 = new SoccerDynamic(); // no argument constructor
        SoccerDynamic soccer2 = new SoccerDynamic("Ronaldo", "Saudi Pro", "ST"); // one argument constructor

        System.out.println("Info:"); // accessing getter
        System.out.println("Player 1:");
        soccer1.displayInfo();
        System.out.println(); 

        System.out.println("Player 2:");
        soccer2.displayInfo();
        System.out.println();
    }
}
// IJava activation
SoccerDynamic.main(null);
```

# Java Console Games

```java
import java.util.Random; //library for generating random.
import java.util.Scanner; //library for the user input

public class CoinFlipGame { //class name 
    public static void main(String[] args) { // start of the program
        Scanner scanner = new Scanner(System.in); // create the scanner object to read input 
        Random random = new Random(); // creates the random object 

        System.out.println("Welcome to the Coin Flip Game!");// display in the console
        System.out.print("Guess the coin toss! Enter 'heads' or 'tails': ");// prompt of question
        String userGuess = scanner.next(); // reads the guess from the console and store it as a guess variable

        //  this is used to Generate a random 0 or 1 for heads or tails
        int coinToss = random.nextInt(2); // 0 represents heads, 1 represents tails

        String result = (coinToss == 0) ? "heads" : "tails"; //this is checking the result if its = to 0 then its heads or 1 its tails 

        if (userGuess.equalsIgnoreCase(result)) { // This is to check if the user guess in userGuess is matches the result 
            System.out.println("Congratulations! It's " + result + ". You win!");
        } else {
            System.out.println("Sorry, it's " + result + ". You lose.");// Doesent Match you get this 
        }
    }
}

// Units Used
//  Unit 1: Primitive Data Types 
// Unit 2 Using Objects 
// unit 3: boolean expressions
```
